```python
import numpy as np #Problem 4

arr4 = np.random.randint(0, 101, (5, 5))
print(arr4)

print(np.max(arr4, axis=1))
print(np.sum(arr4, axis=0))
print(np.mean(arr4))

new_matrix = (arr4 > 50).astype(int)
print(new_matrix)
```
